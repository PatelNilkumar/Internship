```python
product_titles=[]
brand_names=[]
price_list=[]
```

```python
titles_tags=driver.find_elements_by_xpath("//a[@class='IRpwTa']")
titles_tags[0:10]
```

```python
for i in titles_tags:
    title=i.text
    product_titles.append(title)
product_titles[0:10] 
```

```python
brand_tags=driver.find_elements_by_xpath("//div[@class='_2B099V']")
brand_tags[0:10]
```

```python
for i in brand_tags:
    brand_name=i.text
    brand_names.append(brand_name)
brand_names[0:10]
```

```python
price_tags=driver.find_elements_by_xpath("//a[@class='_3bPFwb']")
price_tags[0:10]
```

```python
for i in price_tags:
    price=i.text
    price_list.append(price)
price_list[0:10]   
```
