```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://meesho.com/bags-ladies/pl/p7vbp')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
first_title = soup.find('div',class_="sc-gsnTZi Grid__Row-sc-4ki5nk-0 gfMTeL WphYx")
first_title
```

```python
first_title.text
```

```python
Discount = soup.find('div',class_="Card__BaseCard-sc-b3n78k-0 iKnziG NewProductCard__DiscountRow-sc-j0e7tu-15 fpGaRM NewProductCard__DiscountRow-sc-j0e7tu-15 fpGaRM")
Discount.text
```

```python
sta = soup.find('span',class_="Text__StyledText-sc-oo0kvp-0 dLSsNI")
sta.text.split()[1]
```

```python
titles = [] 

for i in soup.find_all('div',class_="sc-gsnTZi Grid__Row-sc-4ki5nk-0 gfMTeL WphYx"):
    titles.append(i.text)
    
    
titles 
```

```python
Discount = [] # emplty list

for i in soup.find_all('div',class_="Card__BaseCard-sc-b3n78k-0 iKnziG NewProductCard__DiscountRow-sc-j0e7tu-15 fpGaRM NewProductCard__DiscountRow-sc-j0e7tu-15 fpGaRM"):
    Discount.append(i.text)
    
Discount    
```

```python
price = [] # emplty list

for i in soup.find_all('span',class_="Text__StyledText-sc-oo0kvp-0 dLSsNI"):
    price.append(i.text.replace('',''))
    
price  
```

```python
images = []
for i in soup.find_all("img",class_="Voguish Alluring Women Handbags"):
    images.append(i['data-src'])
images 
```
